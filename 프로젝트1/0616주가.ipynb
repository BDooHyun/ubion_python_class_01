```python
import pandas as pd
```

```python
stock = pd.read_csv("./datasets/0616/대한항공주가데이터.csv")
stock = stock.sort_values(by = '날짜', ascending=True).reset_index(drop=True)
stock['변동 %'] = stock['변동 %'].str.rstrip('%').astype(float)
stock['종가'] = stock['종가'].str.replace(',', '').astype(float)
stock = stock.drop(["시가", "고가", "저가", "거래량"], axis=1)
# stock[stock["거래량"].isna()]
```

```python
stock
```

```python
gold = pd.read_csv("./datasets/0616/금데이터.csv")
gold = gold.sort_values(by = '날짜',ascending=True)
gold = gold.reset_index(drop=True)
gold['종가'] = gold['종가'].str.replace(',', '').astype(float)
gold.rename(columns={'종가':'금종가'},inplace=True)
gold = gold[['날짜','금종가']]
# gold[gold["거래량"].isna()]
```

```python
gold
```

```python
gold_krx = pd.read_csv("./datasets/0616/금krx.csv", encoding="euc-kr")
# gold = gold.sort_values(by = '날짜',ascending=True)
# gold = gold.reset_index(drop=True)
# gold['종가'] = gold['종가'].str.replace(',', '').astype(float)
# gold.rename(columns={'종가':'금종가'},inplace=True)
# gold = gold[['날짜','금종가']]
```

```python
gold_krx.head()
```

```python
gold_krx = gold_krx.sort_values(by = '일자',ascending=True).reset_index(drop=True)
```

```python
gold_krx.rename(columns={'일자':"날짜"},inplace=True)
```

```python
gold_krx.rename(columns={'종가':"금종가"},inplace=True)
```

```python
gold_krx = gold_krx[['날짜','금종가']]
```

```python
gold_krx["날짜"] = gold_krx["날짜"].astype(str)
```

```python
gold_krx["날짜"] = gold_krx["날짜"].str.replace('/', '- ')
```

```python
gold_krx
```

```python
oil = pd.read_csv("./datasets/0616/유가데이터.csv")
oil = oil.sort_values(by = '날짜',ascending=True)
oil = oil.reset_index(drop=True)
oil.rename(columns={'종가':'유가종가'},inplace=True)
oil =oil[['날짜','유가종가']]
# oil[oil["거래량"].isna()]
```

```python
oil
```

```python
stock_gold = pd.merge(gold_krx, stock, how='left',on='날짜')
```

```python
stock_gold
```

```python
main_df = pd.merge(stock_gold, oil, how='left',on='날짜')
```

```python
main_df
```

```python
main_df['target']  = main_df['종가'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
main_df.drop("변동 %", axis=1, inplace=True)
```

```python
main_df = main_df[["날짜", "종가", "금종가", "유가종가", "target"]]
```

```python
main_df["날짜"] = main_df["날짜"].str.replace('- ', '-')
```

```python
main_df.to_csv('./주가데이터(0616).csv', index=False)
```
