```python
import pandas as pd
```

```python
dh = pd.read_excel('./테스트데이터/(테스트)대한주가.xlsx')
lg = pd.read_excel('./테스트데이터/(테스트)엘지주가.xlsx')
```

```python
dh = dh[["일자", "종가"]]
lg = lg[["일자", "종가"]]
```

```python
dh.columns = ['일자', '대한종가']
lg.columns = ['일자', '엘지종가']
```

```python
total = pd.merge(dh, lg, how="inner")
```

```python
total.sort_values(by="일자", ignore_index=True, inplace=True)
```

```python
total
```

```python
_standard = total["DL"][0]
```

```python
total["DL"] = (total["DL"] / _standard) * 100
```

```python
total.head(15)
```

```python
total['target']  = total['DL'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
total['대한타겟'] = total['대한종가'].diff().apply(lambda x: 1 if x > 0 else 0)
total['엘지타겟'] = total['엘지종가'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
DH_stock = total[["일자", '대한종가', "대한타겟"]]
LG_stock = total[["일자", '엘지종가', "엘지타겟"]]
```

```python
DH_stock.to_excel("./테스트데이터/(테스트)DH주가타겟.xlsx")
LG_stock.to_excel("./테스트데이터/(테스트)LG주가타겟.xlsx")
```

```python
dh['대한타겟'] = dh['대한종가'].diff().apply(lambda x: 1 if x > 0 else 0)
lg['엘지타겟'] = lg['엘지종가'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
dh.to_excel('./테스트데이터/(테스트)대한주가.xlsx')
lg.to_excel('./테스트데이터/(테스트)엘지주가.xlsx')
```
