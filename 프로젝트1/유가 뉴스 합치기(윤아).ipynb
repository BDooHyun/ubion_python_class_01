```python
oil_news_time_df =pd.read_excel('./3000~4500/유가데이터(시간) 3001~4500.xlsx')
oil_news_time_df = oil_news_time_df.drop(['Unnamed: 0'],axis=1)
```

```python
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('\n','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('\t','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('(','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace(')','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('\r','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('입력','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('기자명','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('지면','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('승인','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('댓글 0','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('수정','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('VIEW','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('등록','')
oil_news_time_df['시간'] = oil_news_time_df['시간'].str.replace('파이낸셜뉴스','')
```

```python
oil_news_time_df.to_excel('./4500시간 수정.xlsx')
```
