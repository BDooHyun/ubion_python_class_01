```python
import pandas as pd
```

```python
dh_df = pd.read_csv('./datasets/대한항공주가데이터.csv')
dh_df = dh_df.sort_values(by = '날짜', ascending=True).reset_index(drop=True)
dh_df.info()
```

```python
dh_df['변동 %'] = dh_df['변동 %'].str.rstrip('%').astype(float)
```

```python
dh_df['종가'] = dh_df['종가'].str.replace(',', '').astype(float)
```

```python
# dh_df['Label']  = dh_df['종가'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
dh_df = dh_df.drop(["시가", "고가", "저가", "거래량"], axis=1)
dh_df
```

```python
gold_df = pd.read_csv('./datasets/금 데이터.csv')
gold_df = gold_df.sort_values(by = '날짜',ascending=True)
```

```python
gold_df = gold_df.reset_index(drop=True)
gold_df
```

```python
gold_df['종가'] = gold_df['종가'].str.replace(',', '').astype(float)
```

```python
gold_df.rename(columns={'종가':'금종가'},inplace=True)
```

```python
gold_df = gold_df[['날짜','금종가']]
gold_df
```

```python
oil_df = pd.read_csv('./datasets/WTI유 선물 과거 데이터.csv')
oil_df = oil_df.sort_values(by = '날짜',ascending=True)
```

```python
oil_df = oil_df.reset_index(drop=True)
oil_df.info()
```

```python
oil_df.rename(columns={'종가':'유가종가'},inplace=True)
```

```python
oil_df =oil_df[['날짜','유가종가']]
```

```python
oil_df
```

```python
main_df = pd.merge(dh_df,gold_df,how='inner',on='날짜')
main_df
```

```python
main_df = pd.merge(main_df,oil_df,how='inner',on='날짜')
```
