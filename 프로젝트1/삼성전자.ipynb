```python
import pandas as pd
```

```python
samsung_df = pd.read_csv('../삼성전자 220101~20230531.csv')
samsung_df = samsung_df.sort_values(by = '날짜',ascending=True).reset_index(drop=True)
samsung_df.info()
```

```python
samsung_df['변동 %'] = samsung_df['변동 %'].str.rstrip('%').astype(float)
```

```python
samsung_df['종가'] = samsung_df['종가'].str.replace(',', '').astype(float)
```

```python
samsung_df['Label']  = samsung_df['종가'].diff().apply(lambda x: 1 if x > 0 else 0)
```

```python
samsung_df
```

```python
gold_df = pd.read_csv('../금 데이터.csv')
gold_df = gold_df.sort_values(by = '날짜',ascending=True)
```

```python
gold_df = gold_df.reset_index(drop=True)
gold_df
```

```python
gold_df['종가'] = gold_df['종가'].str.replace(',', '').astype(float)
```

```python
gold_df.rename(columns={'종가':'금종가'},inplace=True)
```

```python
gold_df = gold_df[['날짜','금종가']]
gold_df
```
