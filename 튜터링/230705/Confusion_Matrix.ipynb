## 1. 혼동행렬

# 예시

```python
import numpy as np

# 주어진 값으로 혼동행렬 생성
A_병원 = np.array([[9, 1], [30, 60]])

print(A_병원)
```

```python
# 주어진 값으로 혼동행렬 생성
B_병원 = np.array([[1, 9], [20, 70]])

print(B_병원)
```

---
## 2. 정확도 (Accuracy)

```python
# Accuracy(정확도) 계산
A_accuracy = (A_병원[0, 0] + A_병원[1, 1]) / np.sum(A_병원)

B_accuracy = (B_병원[0, 0] + B_병원[1, 1]) / np.sum(B_병원)

# 결과 출력
print("A_병원_Accuracy:", A_accuracy)
print("B_병원_Accuracy:", B_accuracy)
```

---
## 3. 정밀도 (Precision)

```python
# precision(정밀도) 계산
A_precision = A_병원[0, 0] / (A_병원[1, 0] + A_병원[0, 0])

B_precision = B_병원[0, 0] / (B_병원[1, 0] + B_병원[0, 0])

# 결과 출력
print("A_병원_precision:", round(A_precision,2))
print("B_병원_precision:", round(B_precision,2))
```

---
## 4. 재현률 (Recall)

```python
# recall(재현률) 계산
A_recall = A_병원[0, 0] / (A_병원[0, 1] + A_병원[0, 0])

B_recall = B_병원[0, 0] / (B_병원[0, 1] + B_병원[0, 0])

# 결과 출력
print("A_병원_recall:", round(A_recall,2))
print("B_병원_recall:", round(B_recall,2))
```

---
## 5. F1_Score

```python
# F1 Score(F1 점수) 계산
A_f1_score = 2 * (A_precision * A_recall) / (A_precision + A_recall)

B_f1_score = 2 * (B_precision * B_recall) / (B_precision + B_recall)

# 결과 출력
print("A_병원_f1_score:", round(A_f1_score,2))
print("B_병원_f1_score:", round(B_f1_score,2))
```

# 부실기업

- Accuracy(정확도): Accuracy는 전체 샘플 중에서 부실기업을 올바르게 예측한 샘플의 비율을 나타냅니다. 정확도는 모든 예측 결과를 종합적으로 평가하며, 높은 정확도는 모델이 부실기업을 식별하는 데 더 잘 수행하고 있다는 것을 의미합니다.

- Precision(정밀도): Precision은 모델이 부실기업으로 예측한 샘플 중에서 실제로 부실한 비율을 나타냅니다. 즉, 모델이 부실기업이라고 분류한 샘플 중에서 실제로 부실한 기업의 비율을 측정합니다. 높은 정밀도는 모델이 부실로 분류한 기업이 실제로도 부실한 기업일 확률이 높다는 것을 나타냅니다.

- Recall(재현율): Recall은 실제로 부실한 기업 중에서 모델이 정확하게 부실로 예측한 비율을 나타냅니다. 즉, 실제로 부실한 기업을 모델이 얼마나 잘 탐지하는지 측정합니다. 높은 재현율은 모델이 부실한 기업을 빠뜨리지 않고 잘 찾아낸다는 것을 의미합니다.

- F1 Score(F1 점수): F1 Score는 Precision과 Recall의 조화 평균으로, 정밀도와 재현율의 균형을 나타냅니다. F1 Score는 Precision과 Recall이 모두 높을 때 가장 높은 값을 가지며, 두 지표 사이의 트레이드오프를 평가하는 데 사용됩니다. F1 Score가 높을수록 모델이 정확하게 부실기업을 예측하는 데 더 효과적이라고 할 수 있습니다.
