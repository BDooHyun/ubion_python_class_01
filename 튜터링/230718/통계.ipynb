# 0. 귀무가설과 대립가설
- 귀무 가설(H0):
    - 귀무 가설은 일반적으로 효과가 없다는 가정이며, 연구자의 기존 생각이나 예상을 반영합니다. 귀무 가설은 검정을 통해 반증하고자 하는 가설로서, 일반적으로 "등호(=)"로 표현됩니다. 
#
- 대립 가설(H1):
    - 대립 가설은 귀무 가설의 반대이며, 연구자가 검정을 통해 입증하고자 하는 가설입니다. 대립 가설은 귀무 가설이 거짓일 때 참이 되는 가설로서, 일반적으로 "부등호(<, >, ≠)"로 표현됩니다.
# 
- 요약
    - 가설 설정은 연구의 목적과 문제에 따라 달라지며, 이를 기반으로 통계적 검정을 수행하여 가설을 받아들일지 기각할지를 결정합니다. 일반적으로 통계 분석에서는 귀무 가설을 기각하고 대립 가설을 채택하는 것이 주된 관심사입니다. 이를 통해 데이터를 통계적으로 분석하고 결론을 도출할 수 있습니다.

# 1. Z-test(Z-검정)
- Z-검정은 통계 분석에서 가설 검정에 사용되는 방법 중 하나로, 주로 모집단의 평균에 대한 가설을 검정하는 데 사용됩니다. 
#
- 한계점
    - 표본 크기에 따른 차이 : Z-test는 표본 크기가 큰 경우에 더 정확한 결과를 제공합니다. 그러나 표본 크기가 작은 경우, 모집단의 분산을 추정하기 어렵기 때문에 T-test가 더 적합합니다
    - 모집단의 분산을 알 수 없는 경우 : Z-test는 모집단의 분산을 정확히 알고 있는 경우에 적용됩니다. 그러나 대부분의 실제 상황에서는 모집단의 분산을 알 수 없기 때문에 T-test를 사용합니다.
    - 정규분포 가정 : Z-test는 모집단이 정규분포를 따른다는 가정하에 사용됩니다. 그러나 T-test는 모집단의 분포가 정규분포를 따르지 않아도 근사적으로 사용할 수 있습니다. T-test는 중심극한정리에 기반하여, 표본의 크기가 충분하면 모집단의 분포에 관계없이 정규분포에 근사합니다.
#
- 요약
    - Z-test는 표본의 크기가 크고, 모집단의 분포가 정규분포를 따른다는 가정을 만족할 때 정확한 결과를 제공합니다. 모집단의 분포에 대한 정보가 없거나 표본의 크기가 작을 경우에는 T-test와 같은 다른 가설 검정 방법을 고려해야 합니다.

```python
import numpy as np
from scipy import stats

# 그룹 1의 표본 데이터
group1 = np.array([23, 25, 28, 21, 26, 24])

# 그룹 2의 표본 데이터
group2 = np.array([27, 29, 30, 28, 26, 25])

# Z-test 수행
_, p_value = stats.ttest_ind(group1, group2)

alpha = 0.05  # 유의 수준 설정

# p-value와 유의 수준을 비교하여 결과 도출
if p_value < alpha:
    print("두 그룹 간의 평균 차이는 통계적으로 유의미합니다.")
else:
    print("두 그룹 간의 평균 차이는 통계적으로 유의미하지 않습니다.")
```

# 2. T-test(T-검정)
- T-검정은 집단 간의 차이를 비교하는 방법(t-검정은 두 그룹 간에 차이가 있는지 확인하는 통계 분석 방법)
#
- 예시
    - 어떤 약을 먹은 사람들과 먹지 않은 사람들의 평균 체온을 비교하고 싶다고 가정해봅시다. 먼저, 약을 먹은 사람들의 체온을 측정하고 평균을 구합니다. 
    - 그리고 약을 먹지 않은 사람들의 체온을 측정하고 이들의 평균을 구합니다. 
    - t-검정은 이 두 평균 간의 차이가 우연히 발생한 것인지 아니면 실제로 의미 있는 차이인지를 알려주는 통계적인 방법입니다.
# 
- t-검정의 두 가지 유형
    - 독립 표본 t-검정 : 두 그룹이 서로 독립된 경우에 사용됩니다. 즉, 약을 먹은 그룹과 먹지 않은 그룹의 사람들이 각각 다른 사람들이라는 것입니다.
    - 대응 표본 t-검정 : 같은 사람들의 두 가지 조건을 비교할 때 사용됩니다. 즉, 동일한 사람들에게 약을 먹기 전과 후의 체온을 측정합니다.
#
- 요약
    - t-검정은 두 그룹 간의 차이를 알아보기 위한 통계적인 방법입니다. 이를 통해 우리는 두 그룹이 유의미하게 다른지 아니면 우연히 다른 것인지를 확인할 수 있습니다.

```python
import scipy.stats as stats

# 약을 먹은 그룹의 체온 데이터
group1 = [36.7, 36.9, 37.1, 36.8, 36.9]

# 약을 먹지 않은 그룹의 체온 데이터
group2 = [36.5, 36.6, 36.4, 36.7, 36.5]

# 독립 표본 t-검정 수행
t_value, p_value = stats.ttest_ind(group1, group2)

# 결과 출력
print("t-value:", round(t_value,5))
print("p-value:", round(p_value,5))
```

# T-test 결과 해석
- t-value
    - t-value는 독립 표본 t-검정에서 계산된 t-통계량을 의미합니다.
    - 이 값은 두 그룹의 평균 차이를 표준 오차로 나눈 것을 나타냅니다.
        - 표준 오차(S.E)란?
        - 추정치의 정확성을 나타내는 지표로, 표본의 변동성을 측정하여 얻어지는 값
        - 표준 오차는 이 표본의 변동성을 측정하여, 그 추정치의 신뢰성을 파악하는 데 사용됩니다. 표본은 모집단을 대표하기 위해 무작위로 선택되기 때문에, 다양한 표본을 추출하면 추정치는 표본마다 다를 수 있습니다. 이러한 추정치의 변동성을 표준 오차로 나타내어, 우리가 얻은 추정치가 얼마나 믿을만한지를 판단할 수 있습니다.
            - 표본이란?
            - 모집단에서 추출한 작은 부분집합
            - ex) 전체 학생들의 성적을 알고 싶다면, 일부 학생들을 랜덤하게 선택하여 그들의 성적을 조사하는 것이 표본
    - t-통계량은 양수 또는 음수가 될 수 있으며, 그 값의 크기와 방향성은 평균 차이의 크기와 방향성을 반영합니다.
# 
- 유의점
    - 해석할 때 주의해야 할 점은 t-통계량 값만으로는 평균 차이의 유의성을 판단할 수 없다는 것입니다.
    - t-통계량의 크기 자체는 평균 차이의 크기를 나타내지만, 이 값이 통계적으로 유의미한지를 판단하기 위해서는 p-값과 함께 고려해야 합니다.
#
- 해석
    - t-통계량이 크면 클수록 두 그룹 간의 평균 차이가 크다는 것을 의미합니다. 
    - 하지만 p-값이 유의수준보다 작지 않다면, 해당 평균 차이는 통계적으로 유의미하지 않을 수 있습니다. 
    - 따라서, t-통계량의 크기를 해석할 때는 p-값과 함께 고려하여 평균 차이의 유의성을 평가하는 것이 중요합니다.
#
- 요약
    - t-value은 두 그룹의 평균 차이를 나타내는 값이며, 그 크기와 방향성을 통해 평균 차이의 크기와 방향성을 파악할 수 있습니다. 하지만 유의성 판단을 위해서는 p-값과 함께 고려해야 합니다.

# T-test 결과 해석2
- t-value (t-통계량):
    - t-value는 t-통계량으로 계산된 값으로서, 두 그룹 간의 평균 차이를 표준 오차로 나눈 것을 나타냅니다.
    - t-value는 양수 또는 음수의 값을 가질 수 있으며, 그 값의 크기와 부호는 평균 차이의 크기와 방향성을 반영합니다.
    - t-value의 절대값이 클수록, 즉 평균 차이의 크기가 크면서 t-value가 큰 경우, 해당 평균 차이가 통계적으로 유의미하다는 증거가 더 강해집니다.
#
- p-value (p-값):
    - p-value는 통계적 가설 검정에서 사용되며, 귀무가설이 참일 때 관측된 결과가 나타날 확률을 의미합니다.
    - p-value는 귀무가설을 "기각"할 수 있는 유의성 임계값과 비교하여 검정 결과를 평가하는 데 사용됩니다.
    - p-value가 유의수준(예: 0.05)보다 작으면, 귀무가설을 기각하고 대립가설을 받아들입니다. 이는 해당 결과가 우연에 의한 것보다는 실제로 의미 있는 차이가 있다는 증거로 간주됩니다.
    - p-value는 0에서 1 사이의 값을 가지며, 작을수록 통계적으로 유의미한 차이가 있다는 의미가 강해집니다. 일반적으로, p-value가 0.05보다 작으면 유의미한 결과로 간주됩니다.
#
- 요약
    - t-value는 평균 차이를 나타내는 값이며, 그 크기와 부호는 평균 차이의 크기와 방향성을 나타냅니다. 
    - p-value는 해당 결과의 유의성을 평가하는데 사용되며, 유의수준과 비교하여 검정 결과를 해석합니다. 
    - t-value와 p-value는 함께 고려하여 평균 차이의 유의성을 평가하는 데 중요한 역할을 합니다.

# 3. F-test(F-검정)
- f-검정은 세 개 이상의 그룹 간에 평균 차이가 있는지를 평가하는 분석 방법
- t-검정과 마찬가지로, f-검정은 여러 그룹 간의 차이를 비교하는 데 사용됩니다.
# 
- 예시
    - 세 가지 다른 교육 방법을 시도하여 수학 시험 성적을 비교하고 싶다고 가정해봅시다. 그러면 세 개의 그룹이 있게 되고, 각 그룹의 평균 점수를 계산합니다.
#
- 해석
    - f-검정은 이러한 그룹들의 평균 차이가 우연에 의한 것인지 아니면 실제로 의미 있는 차이인지를 알려주는 통계적인 방법입니다.
    - f-검정은 분산(데이터의 흩어짐 정도)을 이용하여 분석을 수행합니다. 그룹 간에 분산이 유의미하게 다를 때, 즉 그룹 간의 차이가 큰 경우에는 f-검정이 유의미한 결과를 도출합니다.
#
- 요약
    - 최종 목표는 각 집단의 평균을 비교하는 것이고,단지 그 방법을 분산을 이용해서 하는 것입니다.

```python
import scipy.stats as stats

# 교육 방법 A의 수학 시험 성적
method_a = [80, 85, 82, 88, 87]

# 교육 방법 B의 수학 시험 성적
method_b = [75, 78, 80, 85, 82]

# 교육 방법 C의 수학 시험 성적
method_c = [70, 72, 75, 78, 80]

# 분산분석 (ANOVA) 수행
f_value, p_value = stats.f_oneway(method_a, method_b, method_c)

# 결과 출력
print("F-value:", round(f_value,5))
print("p-value:", round(p_value,5))
```

# ANOVA(분산분석)
- ANOVA는 그룹 간의 평균 차이를 평가할 때 그룹 내의 분산과 그룹 간의 분산을 함께 고려하는 것이 특징입니다.
# 
- 분산이란?
    - 분산은 데이터가 평균에서 얼마나 멀리 흩어져 있는지를 나타내는 지표입니다. 그룹 내의 분산은 같은 그룹 내에서 데이터가 얼마나 흩어져 있는지를 나타내고, 그룹 간의 분산은 그룹 간에 평균 차이가 얼마나 큰지를 나타냅니다.
# 
- 예시
    - 세 가지 다른 토마토 품종(A, B, C)이 있다고 합시다. 이 품종들의 수확량을 비교하고 싶다면, 즉 어떤 품종이 가장 많은 수확량을 가지고 있는지 알고 싶다면, 분산분석(ANOVA)을 사용할 수 있습니다.
    - 먼저, 각 품종에서 몇 개의 토마토를 수확했는지 데이터를 수집합니다. 그리고 각 품종의 평균 수확량을 계산합니다.
# 
- 해석
    - ANOVA는 그룹 간의 평균 차이를 평가할 때, 그룹 간의 분산과 그룹 내의 분산을 비교합니다.
    - 그룹 간의 분산이 크면서 동시에 그룹 내의 분산이 작을수록 ANOVA는 유의미한 결과를 제시합니다.
    - 그룹 간의 분산이 크면 그룹 간의 평균 차이가 커진다는 것을 의미합니다. 즉, 그룹들이 평균적으로 서로 다른 값을 가지면 그룹 간의 분산이 커질 것입니다.
    - 이와 동시에 그룹 내의 분산이 작을수록, 즉 그룹 내 데이터들이 비교적 유사한 값을 가질수록 ANOVA는 유의미한 결과를 나타냅니다.
#
- 요약
    - ANOVA는 그룹 간의 평균 차이를 평가할 때 그룹 간의 분산과 그룹 내의 분산을 함께 고려합니다. 
    - 그룹 간에 평균의 차이가 우연에 의한 것인지 아니면 실제로 의미 있는 차이인지를 알려주는 통계적인 방법입니다.
    - 그룹 간의 분산이 크고 그룹 내의 분산이 작을수록 ANOVA는 유의미한 결과를 제시하며, 이는 그룹 간의 평균 차이가 실제로 의미 있는 차이일 가능성이 크다는 것을 의미합니다.
#
- 한계점
    - ANOVA 결과는 유의미한 평균 차이가 있다는 것을 보여주는 것뿐이며, 어느 그룹 간에 평균이 다른지 구체적으로 알려주지는 않습니다.
    - 따라서, 유의미한 결과를 얻은 경우, 추가적인 다중 비교 분석이나 사후 검정을 수행하여 어떤 그룹 간에 평균 차이가 있는지를 확인하는 것이 좋습니다.

```python
import scipy.stats as stats

# 품종 A의 수확량 데이터
harvest_a = [120, 125, 130, 135, 140]

# 품종 B의 수확량 데이터
harvest_b = [115, 118, 122, 128, 130]

# 품종 C의 수확량 데이터
harvest_c = [105, 110, 115, 118, 122]

# harvest_c = [115, 117, 120, 128, 130]

# 분산분석 (ANOVA) 수행
f_vlaue, p_value = stats.f_oneway(harvest_a, harvest_b, harvest_c)

# 결과 출력
print("F-value:", round(f_value,5))
print("p-value:", round(p_value,5))
```

# ANOVA와 F-검정
- ANOVA(Analysis of Variance)와 F-검정은 서로 연관된 개념입니다. 간단히 말하자면, ANOVA는 F-검정을 사용하여 그룹 간의 평균 차이를 평가하는 통계 분석 방법입니다.
- ANOVA는 세 개 이상의 그룹 간의 평균 차이를 검정하는 분석 방법을 의미합니다. 이때, 그룹 간의 평균 차이가 통계적으로 유의미한지를 판단하기 위해 F-검정을 수행합니다. 따라서, F-검정은 ANOVA에서 사용되는 통계적인 기법 중 하나입니다.
- F-검정은 그룹 간의 분산 비율을 비교하여 평균 차이의 유의성을 판단하는데 사용됩니다. 그룹 간의 분산과 그룹 내의 분산을 이용하여 계산된 F-값은 그룹 간의 평균 차이에 대한 측정 지표입니다. F-검정은 그룹 간의 평균 차이가 우연에 의한 것인지 아니면 실제로 의미 있는 차이인지를 판단합니다.
#
- 요약
    - ANOVA는 세 개 이상의 그룹 간의 평균 차이를 검정하는 분석 방법이고, F-검정은 ANOVA에서 그룹 간의 평균 차이를 평가하기 위해 사용되는 통계적인 기법입니다.
